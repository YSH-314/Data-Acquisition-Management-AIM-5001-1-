## Part 1: Regular Expressions (30 Points)
Text data is often in need of “cleaning” and preparation before it can be effectively used for analysis purposes. Consider the following poorly formatted text string containing names and phone numbers of some residents of the town of Springfield:

__"555-1239Dr. Anthony Fauci(636) 555-0113Hollingdorp, Donnatella555-6542Fitzgerald, F. Scott555 8904Rev. Martin Luther King636-555-3226Snodgrass, Theodore5553642Carlamina Scarfoni"__

Use your Python regular expression (“regex”) skills to complete the following tasks:


### 1. (6 Points) Extract the names of each individual from the unformatted text string shown above and store them in a vector of some sort. When complete, your vector should contain the following entries:

"Dr. Anthony Fauci" "Hollingdorp, Donnatella" "Fitzgerald, F. Scott" "Rev. Martin Luther King" "Snodgrass, Theodore" "Carlamina Scarfoni"

```python
import re
```

```python
# to extract all the names in the text. re.findall will find all the matched names and put them into a list 
text="555-1239Dr. Anthony Fauci(636) 555-0113Hollingdorp, Donnatella555-6542Fitzgerald, F. Scott555 8904Rev. Martin Luther King636-555-3226Snodgrass, Theodore5553642Carlamina Scarfoni"
match_all=(re.findall('[a-zA-Z\.,\s]+[a-zA-Z\.]+',text))
print(match_all)
```

### 2. Using your new vector containing only the names of the six individuals, complete the following tasks:  

### a. (4 Points) Use your regex skills to rearrange the vector so that all elements conform to the standard “firstname lastname”, preserving any titles (e.g., “Rev.”, “Dr.”, etc) or middle/second names.

```python
# Use the result from Q1, match_all, to rearrange. 
for i in range(len(match_all)):
    match=re.search('([a-zA-Z]+), ([a-zA-Z\.\s]+)',match_all[i])
    if match: # means if match != None
        last, fisrt=match.group(1),match.group(2)
        #print(type(last)) 
        match_all[i]=fisrt+' '+last
print(match_all)
```

### b. (4 Points) Using your regex skills, construct a logical vector indicating whether a character has a title (i.e., Rev. and Dr.).

```python
# The format of title is a capital letter followed by a few small letters and a dot.
logical_vector=[]
for i in range(len(match_all)):
    logical_vector.append(bool(re.search('[A-Z][a-z]+[\.]',match_all[i])))
print('whether a character has a title:',logical_vector)
```

### c. (4 Points) Using your regex skills, construct a logical vector indicating whether a character has a middle/second name.

```python
# Utilize the number of spaces to find whether have a middle name or not.
logical_vector=[]
for i in range(len(match_all)):
    match=bool(re.search('[A-Z][a-z]+[\.]',match_all[i])) 
    if not match: # if the name did not contain title, then there should have 2 spaces during a string of character
        logical_vector.append(bool(re.search('.+\s.+\s.+',match_all[i])))
    else:            # if the name contained title, then there should have 3 spaces during a string of character
        logical_vector.append(bool(re.search('.+\s.+\s.+\s.+',match_all[i])))
print('whether a character has a middle/second name:',logical_vector)
```

### 3. (6 Points) Consider the HTML string \<title>+++BREAKING NEWS+++\<title>. We would like to extract the first HTML tag (i.e., “\<title>”). To do so we write the regular expression “<.+>”. Explain why this fails and correct the expression.

__Ans: Since the "." in “<.+>” means every character, and letters (a-z or A-Z) and signs ("<" , ">" , "+") in this case are all characters, we will extract the whole length of the string when search with “<.+>”. By substituting "." with [a-z] and with "^" in the first place will extract the first HTML tag since we restrict it to search only small letters that start with "<" and end with ">". 
The code is as the following.__

```python
string='<title>+++BREAKING NEWS+++<title>.'
print(re.search('^<[a-z]+>',string))
```

### 4. (6 Points) Consider the string “(5-3)^2=5^2-2\*5\*3+3^2” conforms to the binomial theorem. We would like to extract the equation in its entirety from the string. To do so we write the regular expression “[^0-9=+*()]+”. Explain why this fails and correct the expression.

__Ans: Some characters have specific function in regex. Ex: "^":search string starts with; "-" in 0-9: represent a range; "*":search string with zero or more occurrences;"()": capture and group. As a result, we need to add "\\" before these characters to escape the function. For instance: "\\^": will search for the character "^", not for searching string start with. To extract the whole equation, the code is as the following.__

```python
string='(5-3)^2=5^2-2*5*3+3^2'
print(re.search('[\^0-9=+\*\(\)-]+',string))
```

## Part 2 - Attempt 1
Attempt 1 utilized lists to store the content of the table. We extract the information with for loop. 

```python
import pandas as pd
```

```python
#read the file and store the line by line into a list
with open('Project1.txt') as f:
    lines = f.readlines()
lines = [line.rstrip() for line in lines if '--' not in line] # skip the dashed line 
for i in lines:
    print(i)
    print()
```

```python
# extract names, total point, opponent from the line with even number
names=[]
total_point=[]
opponent_1=[]
for i in range(2,len(lines),2):
    match=re.search('\| ([A-Z ]+\-?[A-Z]+) +\|',lines[i]) 
    match_2=re.search('\|(\d\.\d)',lines[i])
    match_3=re.findall('\|[A-Z].\s+(\d+)',lines[i])
    
    names.append(match.group(1))
    total_point.append(match_2.group(1))
    opponent_1.append(match_3)
        
        
#print(len(names))
```

```python
# extract state, pre-rating, opponent from line with odd number
state=[]
pre_rating=[]
opponent_2=[]
for i in range(3,len(lines),2):
    #print(lines[i])
    match=re.search('[A-Z]{2}',lines[i]) 
    match_2=re.search('R:[\s]+([\dA-Z]+)',lines[i])
    match_3=re.findall('\|[A-Z].\s+(\d+)',lines[i])
    
    state.append(match.group(0))
    pre_rating.append(match_2.group(1))
    opponent_2.append(match_3)
        
        
```

```python
#combine the opponents in all lines. all_opponent is a list contains 64 lists storing the No. of opponents for each player 
all_opponent=[]
for i in range(len(opponent_1)):
    if len(opponent_2)==0:
        all_opponent.append([int(g) for g in opponent_1[i]]) # transfer the number from string to interger
    else:
        lsall=opponent_1+opponent_2
        all_opponent.append([int(g) for g in lsall[i]]) # transfer the number from string to interger
#print(len(all_opponent))
```

```python
# create a dictionary to retrieve the pre-rating for each player. key:the number of player；value: the pre-rating number
dic_prerating={}
for i in range(len(pre_rating)):
    dic_prerating[i+1]=pre_rating[i] # the number of player starts with 1
```

```python
# calculate average rating of opponents for each player
ave=[]
for i in range(len(all_opponent)):
    rating=[] # store the pre-ratings of each opponent for each player
    for opponent in all_opponent[i]:
        match=re.search('[A-Z]+',dic_prerating[opponent]) # extract the first 4 digits of pre-rating, which has letters in
        if match:
            rating.append(int(dic_prerating[opponent][:match.span()[0]]))
        else:
            rating.append(int(dic_prerating[opponent]))
    #print(o_rating)
    ave.append(int(sum(rating)/len(rating)))      
```

```python
data=pd.DataFrame(list(zip(names,state,total_point,pre_rating,ave)),
                  columns=['Name','state','total points',
                           'pre-tournament rating','Average Pre Tournament Chess Rating of Opponents'])
data.to_csv('final_result.csv')
```
